```python
import anndata
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.supervised import anndata_to_supervised
```

```python
# Generate a fake anndata object

# Some hyperparameters
NUM_SUBJECTS = 10
MIN_SAMPLES = 4
MAX_SAMPLES = 10
NUM_FEATURES = 20

# Initialize subject IDs and sample counts
subject_ids = list(range(NUM_SUBJECTS))
n_time_points = np.random.randint(
    low=MIN_SAMPLES, high=MAX_SAMPLES, size=len(subject_ids)
)

# Some random regression coefficients
beta = np.random.normal(
    loc=np.zeros(NUM_FEATURES), scale=1.0, size=(NUM_FEATURES, NUM_FEATURES)
)
time_coef = np.random.rand()

# Initialize a complicated but blank dataframe
data = pd.DataFrame(
    columns=list(range(NUM_FEATURES)),
)
metadata = pd.DataFrame(columns=["subject", "time"])

# Fill in the dataframe
for subject, n_time_points_subject in zip(subject_ids, n_time_points):
    # Set initial time point randomly
    time = np.random.randint(low=0, high=100)
    val = np.abs(
        np.random.normal(
            loc=np.zeros(NUM_FEATURES), scale=1.0, size=NUM_FEATURES
        )
    )
    data.loc[len(data)] = val
    metadata.loc[len(metadata)] = [subject, time]
    for i in range(1, n_time_points_subject):
        # Generate next time point
        new_time = np.random.randint(low=time, high=100)
        time_diff = new_time - time
        time = new_time
        new_val = np.abs(time_coef * time_diff + val @ beta)
        data.loc[len(data)] = new_val
        val = new_val
        metadata.loc[len(metadata)] = [subject, time]
```

```python
# Normalize the values

data_normalized = data.copy()
data_normalized /= data_normalized.sum(axis=1).values.reshape(-1, 1)
data_normalized
```
